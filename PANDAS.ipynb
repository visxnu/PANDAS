# Data Cleaning

## Dropna Method

```python
df = pd.read_csv('data.csv')
print(df)
```

```python
df = pd.read_csv('data.csv')

new_df = df.dropna()

print(new_df.to_string())
```

```python
df = pd.read_csv('data.csv')

df.dropna(inplace = True)

print(df.to_string())
```

## Fillna Method

```python
df = pd.read_csv('data.csv')
df.fillna(100,inplace=True)
print(df.to_string())
```

```python
df = pd.read_csv('company.csv')
df['experience'].fillna(10,inplace=True)
df
```

### Filling Null Value with Mean

```python
df=(pd.read_csv('company.csv'))
x=df["experience"].mean()
df["experience"].fillna(x,inplace=True)
df
```

```python
df=(pd.read_csv('data.csv'))
x=df['Calories'].mean()
df['Calories'].fillna(x,inplace=True)
print(df.to_string())
```

### Filling Null value with Median

```python
df=(pd.read_csv('data.csv'))
x=df['Calories'].median()
df['Calories'].fillna(x,inplace=True)
print(df.to_string())
```

```python
df=(pd.read_csv('company.csv'))
x=df["experience"].median()
df["experience"].fillna(x,inplace=True)
df
```

### Filling Null value with Mode

```python
df=(pd.read_csv('data.csv'))
x=df['Calories'].mode()[0]
df['Calories'].fillna(x,inplace=True)
print(df.to_string())
```

```python
df=(pd.read_csv('company.csv'))
x=df["experience"].mode()[0]
df["experience"].fillna(x,inplace=True)
df
```

## Handleing Wrong Data

### Replaceing Data

```python
data = {
    'name':["Vishnu","Akhil","Amrutha","Athulya","Ajith","Surya","Manu"],
    'age':[26,18,24,19,23,20,22],
    'salary':[240000,13000,20000,25000,15000,18000,24000]

}
```

```python
df2=pd.DataFrame(data)
df2
```

```python
df2.loc[0,'salary'] = 24000
print(df2)
```

### Replaceing data using for loop

```python
data = {
    'name':["Vishnu","Akhil","Amrutha","Athulya","Ajith","Surya","Manu"],
    'age':[26,18,24,19,23,20,22],
    'salary':[240000,13000,20000,25000,15000,18000,24000]

}
df=pd.DataFrame(data)
```

```python
for i in df.index:
  if df.loc[i,'salary'] > 24000:
    df.loc[i,'salary'] = 20000

df
```

```python
df.loc[2,"salary"] = 75000
df
```

```python
for i in df.index:
  if df.loc[i,'salary'] < 20000:
    df.loc[i,'salary'] = 45000

df
```

```python
for i in df.index:
  if df.loc[i,'age'] > 20:
    df.loc[i,'salary'] = 50000

df
```

## Remove Rows

```python
for j in df.index:
  if df.loc[j,"salary"] < 25000:
    df.drop(j, inplace = True)
```

```python
print(df)
```

# Plotting

### Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data.csv')

df.plot()

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Calories'],color='red',marker='^')
plt.title('Line')
plt.xlabel('Index')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```

### Scatter plot

```python
plt.figure(figsize=(8,5))
plt.scatter(df.index,df['Calories'],color='red',marker='*')
plt.title('Line')
plt.xlabel('Index')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```
