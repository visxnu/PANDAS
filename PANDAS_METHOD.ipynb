## agg () Method [aggregate method]

```python
import pandas as pd
data = {
    'x' : [50,40,30],
    'y' : [300,1112,42]
}

df=pd.DataFrame(data)

x = df.agg(["sum"],axis=1)
x
```

```python
import pandas as pd
data = {
    'x' : [50,40,30],
    'y' : [300,1112,42]
}

df=pd.DataFrame(data)

x = df.agg(["count","sum","min","max"],axis=1)
x
```

## combine() Methods

```python
import pandas as pd

df1 = pd.DataFrame([[1, 2], [3, 4]])
df2 = pd.DataFrame([[5, 6], [7, 8]])

def myfunc(a, b) :
  if (a.sum() > b.sum()):
    return a
  else:
    return b
print(df1)
print(df2)
df1.combine(df2, myfunc)
```

```python
import pandas as pd

df1 = pd.DataFrame([[1,2,3], [13,4,3]])
df2 = pd.DataFrame([[5, 6], [7, 8]])

def myfunc(a, b) :
  if (a.sum() > b.sum()):
    return a
  else:
    return b
print(df1)
print(df2)
df1.combine(df2, myfunc)
```
