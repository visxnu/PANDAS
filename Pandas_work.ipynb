2.You have the following DataFrame:

data = {&#39;Name&#39;: [&#39;Alice&#39;, &#39;Bob&#39;, &#39;Charlie&#39;, &#39;David&#39;],&#39;Age&#39;: [25, 30, None, 40],&#39;City&#39;: [&#39;New York&#39;, None, &#39;Chicago&#39;, &#39;San Francisco&#39;],&#39;Salary&#39;: [None, 50000, 60000, None]}

df = pd.DataFrame(data)

a) Replace missing Age values with the mean Age.

b) Replace missing City values with &#39;Unknown&#39;.

c) Drop rows with missing Salary.

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, None, 40],
        'City': ['New York', None, 'Chicago', 'San Francisco'],
        'Salary': [None, 50000, 60000, None]
        }

df = pd.DataFrame(data)
#a
x=df['Age'].mean()
df['Age'].fillna(x,inplace=True)
#b
df['City'].fillna('Unknow',inplace=True)
#c
m=df.dropna()
print(df)
print()
print(m)
```
